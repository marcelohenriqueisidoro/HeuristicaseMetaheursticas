```python
import matplotlib.pyplot as plt

import math

xlabel = []
ylabel = []

xlabel.append(0.01)
ylabel.append(math.exp(-1/0.01))

for i in range (1,51):
  xlabel.append(i)
  ylabel.append(math.exp(-1/i))


plt.plot(xlabel,ylabel)
plt.show()
```

```python
import random
import time

def função(x):
  return (x**2- math.cos(4*math.pi*x))

inicio = time.time()
xlabel = []
ylabel = []

lim_inf = -5
lim_sup = 5

r = random.random()

x = (lim_sup * r + lim_inf*(1-r))

x_star = x

f_star = função(x)


print(f'X inicial = {x_star:.2f}')
print(f'Objetivo inicial = {f_star:.2f}')



SAmax = 100
Temp = 50
alpha = 0.9
iter = 0

while Temp > 0.01:
  while iter < SAmax:
    iter = iter + 1
    nx = x + random.gauss(0,1)
    delta = função(nx) - função(x)
    if delta < 0:
      x = nx
      xlabel.append(time.time() - inicio)
      ylabel.append(x)
      if função(x) < f_star:
        f_star = função(x)
        x_star = x
    else:
      r = random.random()
      if r < math.exp(-delta/Temp):
        x = nx
  iter = 0
  Temp = Temp*alpha

print(f'X final = {x_star:.2f}')
print(f'Objetivo final = {f_star:.2f}')

plt.plot(xlabel,ylabel)
plt.show()
```

Simulated Annealing para o problema da mochila

```python
def gera_sol(sol):
  for i in range(len(sol)):
    sol[i] = random.randint(0,1)
  return sol

sol = [0]*8

x = gera_sol(sol)
print(x)
```
